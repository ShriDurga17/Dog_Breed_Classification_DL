```python
import numpy as np
import tensorflow as tf
#!pip install tensorflow-hub
# !pip install tensorflow-datasets
import tensorflow_hub as hub
import tensorflow_datasets as tfds
```

```python
train_data,validation_data,test_data=tfds.load(name="imdb_reviews",split=('train[:60%]','train[60%:]','test'),as_supervised=True)
```

```python
train_data
```

```python
train_example_batch,train_labels_batch = next(iter(train_data.batch(10)))
```

```python
train_example_batch
```

```python
train_labels_batch
```

```python
embedding="https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1"
```

```python
heb_layer=hub.KerasLayer(embedding,input_shape=[],dtype=tf.string,trainable=True)
```

```python
heb_layer(train_example_batch[:3])
```

```python
model=tf.keras.Sequential()
model.add(heb_layer)
model.add(tf.keras.layers.Dense(16,activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
history = model.fit(train_data.shuffle(10000).batch(100),epochs = 25,validation_data = validation_data.batch(100) ,verbose=1)
```

```python
results=model.evaluate(test_data.batch(100),verbose=2)
for name,value in zip(model.metrics_names,results):
  print("%s: %.3f"%(name,value))
```
